## The above .data file doesn't have headers so I manually assigned column names based on the metadata provided in the UCI repository

```python
columns = [
    "ID", "Diagnosis",
    "radius_mean", "texture_mean", "perimeter_mean", "area_mean", "smoothness_mean",
    "compactness_mean", "concavity_mean", "concave_points_mean", "symmetry_mean", "fractal_dimension_mean",
    "radius_se", "texture_se", "perimeter_se", "area_se", "smoothness_se",
    "compactness_se", "concavity_se", "concave_points_se", "symmetry_se", "fractal_dimension_se",
    "radius_worst", "texture_worst", "perimeter_worst", "area_worst", "smoothness_worst",
    "compactness_worst", "concavity_worst", "concave_points_worst", "symmetry_worst", "fractal_dimension_worst"
]

# Load the .data file
file_path = "wdbc.data"  # Update with the actual file path
breast_df_2 = pd.read_csv(file_path, header=None, names=columns)
```

## The dataset now has id, target variable 'Diagnosis' and 30 numeric features. We can drop 'id', encode target variable into binary and check for missing values

```python
breast_df_3 = breast_df_2.drop(columns=["ID"])
```

```python
breast_df_3["Diagnosis"] = breast_df_3["Diagnosis"].map({"M": 1, "B": 0})  # Malignant = 1, Benign = 0
```

## Exploratory Data Analysis

### 1. Visualize Target variable distribution

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
target_counts= breast_df_3['Diagnosis'].value_counts()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x=target_counts.index,y=target_counts.values, palette='viridis')
plt.title
plt.xlabel
plt.ylabel
plt.xticks([0,1],["Benign","Malignant"])
plt.show()
```

## We observe that there are more benign cases than malignant cases from the bar heights

### 2. Pair plots and scatter plots

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select a subset of features for the pair plot to avoid clutter
subset_features = [
    "radius_mean", "texture_mean", "perimeter_mean", "area_mean", "smoothness_mean", "Diagnosis"
]

#data["Diagnosis"] = data["Diagnosis"].astype("category")

sns.pairplot(
    breast_df_3[subset_features],
    hue="Diagnosis",
    palette={1: "orange", 0: "skyblue"},  # Mapping 'M' (Malignant) and 'B' (Benign)
    diag_kind="kde"
)

plt.suptitle("Pair Plot of Selected Features", y=1.02, fontsize=16)
plt.show()

#####################################################
# Create pair plot
#sns.pairplot(data[subset_features], hue="Diagnosis", palette={0: "skyblue", 1: "orange"}, diag_kind="kde")
#plt.suptitle("Pair Plot of Selected Features", y=1.02, fontsize=16)
#plt.show()
```

```python
# Scatter plot for two features
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=breast_df_3, 
    x="radius_mean", 
    y="texture_mean", 
    hue="Diagnosis", 
    palette={0: "skyblue", 1: "orange"}, 
    alpha=0.8
)
plt.title("Scatter Plot of Radius Mean vs Texture Mean", fontsize=16)
plt.xlabel("Radius Mean", fontsize=12)
plt.ylabel("Texture Mean", fontsize=12)
plt.legend(title="Diagnosis", labels=["Benign (0)", "Malignant (1)"])
plt.show()
```

### Correlation matrix to analyse feature-target relationships

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Encode 'Diagnosis' as numeric for correlation (M -> 1, B -> 0)
#breast_df_3['Diagnosis_num'] = breast_df_3['Diagnosis'].map({'M': 1, 'B': 0})

# Compute correlations
correlation_matrix = breast_df_3.corr()

# Plot a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# Check the correlation of features with the target
correlations = correlation_matrix['Diagnosis'].sort_values(ascending=False)
print(correlations)
```

```python
# Pair plot for highly correlated features
selected_features = ['radius_mean', 'perimeter_mean', 'area_mean']
sns.pairplot(breast_df_3[selected_features], diag_kind='kde')
plt.show()
```

### Based on the above correlation heatmap, correlation matrix values and reducing redundancy by selecting only one feature out of highly correlated features

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Selected features
selected_features = [
    "radius_mean",
    "radius_worst",
    "concave_points_mean",
    "concave_points_worst",
    "concavity_mean",
    "compactness_mean",
]

# Scatter plots for selected features
plt.figure(figsize=(16, 10))
for i, feature in enumerate(selected_features, 1):
    plt.subplot(2, 3, i)
    sns.scatterplot(
        x=feature,
        y="Diagnosis",
        data=breast_df_3,
        hue="Diagnosis",
        palette={1: "red", 0: "blue"},
        alpha=0.7,
    )
    plt.title(f"Scatter Plot: {feature}")
    plt.xlabel(feature)
    plt.ylabel("Diagnosis")
plt.tight_layout()
plt.show()

# Box plots for selected features
plt.figure(figsize=(16, 10))
for i, feature in enumerate(selected_features, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(
        x="Diagnosis",
        y=feature,
        data=breast_df_3,
        palette={1: "red", 0: "blue"}
    )
    plt.title(f"Box Plot: {feature}")
    plt.xlabel("Diagnosis")
    plt.ylabel(feature)
plt.tight_layout()
plt.show()
```
